
# Q8. Handle Missing Values

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman", None, "Neha"]
Marks = [80, 75, None]
```

Fill missing values using `fillna()`.

## Sample Output
```python
Missing values replaced
```


````python
data = pd.DataFrame({
    'Name':['Aman',None,'Neha'],
    'Marks':[80,75,None]
})
print(data)
````

````python
data['Name'].fillna('Unknown',inplace = True)
data['Marks'].fillna(0,inplace = True)
print(data)
````


# Q10. Remove Duplicate Records

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman","Rahul","Aman","Neha"]
```

Check duplicate rows using `duplicated()` and remove them using `drop_duplicates()`.

## Sample Output
```python
Aman
Rahul
Neha
```


````python
data = pd.DataFrame({
    'Name' : ["Aman","Rahul","Aman","Neha"]
})
print(data)
print(data.duplicated())
print(data.drop_duplicates())
````
