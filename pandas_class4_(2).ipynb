# 🐼 Pandas - Class 4: Data Selection & Filtering
Welcome to **Class 4** of our Pandas series. Today we’ll explore techniques to select and filter data in a DataFrame.

## 1. Boolean Indexing & Conditional Selection
- Use comparison operators to build conditions (e.g., `df['col'] > 50`).
- Pass the condition inside `df[...]` to get matching rows.
- Combine multiple conditions with `&` (AND), `|` (OR), and `~` (NOT).
- Parentheses are required when combining conditions.

```python
import pandas as pd
a = {
    'Name' : ['Devraj','Bob','Charlie','David','Emma'],
    'Age' : [25,30,35,40,22],
    'Score' : [80,92,79,85,95]
}
b = pd.DataFrame(a)

#1. select single column
b[b['Age']>27]

#2. Selecting single column score > 85
b[b['Score'] > 85]

#3. multiple condition (AND): Age > 27 AND Score  > 85
b[(b['Age']>27) & (b['Score'] > 85)]

#4. Multiple Condition (OR):Age > 27 (OR) Score > 85
b[(b['Age'] > 27) | (b['Score'] > 85)]

#5. Not(~)
b[~(b['Score'] > 85)]

#6. Age > 22 and Score > 85
b[(b['Age'] > 27 ) & (b['Score'] > 85)]
```

## 2. Filtering Rows with Multiple Conditions
- Combine two or more conditions to refine selection.
- Remember to enclose each condition in parentheses.

```python
#1. Age > 25 AND Score > 85
b[(b['Age'] > 25) & (b['Score'] > 85)]
#2. Age < 35 OR Score >= 90
b[(b['Age'] < 35) | (b['Score'] >= 90)]
#3. Age between 25 and 35 (inclusive)
b[(b['Age'] >= 25) & (b['Age'] <= 35)]
```

## 3. Using `query()` for Cleaner Filtering
- `query()` allows filtering rows with a string expression.
- It's concise and readable for complex conditions.
- Variables can be passed using the `@` symbol.

```python
#1 . simple Query : Age > 25
print(b.query("Age > 25"))

#2. Multiple conditions using Qurey: Age < 35 and Score >= 85
print(b.query("Age < 35 and Score >= 85"))

#3. Multiple condition using Query :Age  >= 30 and score >= 90
print(b.query('Age >= 30 and Score >= 90'))

#4. Threshold (@)
Score = 90 # variable
b.query("Score >= @Score")
```

## 4. String Operations on Columns
- Use the `.str` accessor for vectorized string methods.
- Common methods: `contains`, `replace`, `startswith`, `endswith`, `lower`, `upper`.
- Useful for cleaning or filtering text data.

```python
print(b)
#1. Covert name to lowercase().
print(b['Name'].str.lower())

#2. Covert name to uppercase()
print(b['Name'].str.upper())

#3. Title() -> Captialize first letter of all the words.
print(b['Name'].str.title())

#4. Replace a substring ('Oldstring','Newstring)
b['Name'].str.replace('Charlie','Charles')

#5. Check if name column contain the letter 'a'
b[b['Name'].str.contains('a')]

#6. Startswith['letter']
b[b['Name'].str.startswith('D')]
#7. endswith['letter']
b[b['Name'].str.endswith('e')]
```
