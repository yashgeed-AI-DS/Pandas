## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
import pandas as pd
data = pd.DataFrame({
    'Name':['John','Robert','Harry','Steve'],
    'Age':[25,18,30,22],
    'Marks':[70,95,80,60],
})
print(data)
data.sort_values(by='Age')
```

## Q2. Sort Marks in Descending Order

Input:
Marks = [70, 95, 80, 60]

Output:
95
80
70
60

```python
print(data.sort_values(by = 'Marks', ascending = False))
```

## Q8. Count City Frequency

Input:
City = ['Bhopal','Indore','Bhopal','Jabalpur']

Output:
Bhopal      2
Indore      1
Jabalpur    1

```python
data = pd.DataFrame({
    'City':['Bhopal','Indore','Bhopal','Jabalpur']
})
print(data)
print(data['City'].value_counts())
```

## Q9. Find Percentage Frequency

Input:
City = ['A','A','B','C']

Output:
A = 50%
B = 25%
C = 25%

```python
data = pd.DataFrame({
    'City':['A','A','B','C']
})
print(data)
print(data['City'].value_counts(normalize = True)*100)
```

## Q11. Find Correlation

Input:
Math = [10,20,30,40]
Science = [15,25,35,45]

Output:
1.0

```python
data = pd.DataFrame({
    'Math':[10,20,30,40],
    'Science':[15,25,35,45]
})
print(data)
print(data.corr())
```

## Q12. Find Covariance

Input:
Math = [10,20,30,40]
Science = [15,25,35,45]

Output:
166.67 (approx)

```python
data = pd.DataFrame({
    'Math':[10,20,30,40],
    'Science':[15,25,35,45]
})
print(data)
print(data.cov())
```

## Q14. Student Statistics

Input:
Marks = [60,70,80,90]

Output:
Mean = 75
Median = 75
Max = 90
Min = 60

```python
data = pd.DataFrame({
    'Marks':[60,70,80,90],
})
print(data)
print(data.mean(numeric_only = True))
print(data.median(numeric_only = True))
print(data.max(numeric_only = True))
print(data.min(numeric_only = True))
```

## Q15. Mini Practice

Create a DataFrame with:
Name, Age, Score, Height

Tasks:
1. Sort by Score (Descending)
2. Find Mean Age
3. Find Median Height
4. Show Score Frequency

```python
data = pd.DataFrame({
    'Name':['John','Robert','Harry','Steve'],
    'Age':[25,18,30,22],
    'Score':[70,95,80,60],
    'Height':[170,165,175,180]
})
print(data)
print(data.sort_values(by = 'Score',ascending = False))
print(data['Age'].mean())
print(data['Height'].median())
print(data ['Score'].value_counts())
```
