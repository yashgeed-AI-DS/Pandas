## 1 What is Pandas and Why Use It?
Pandas is a powerful **Python library** for data analysis and manipulation.

Key points:
- Provides two main data structures: **Series** (1D) and **DataFrame** (2D).
- Makes working with tabular data easy and fast.
- Ideal for cleaning, transforming, and analyzing datasets.
- Works well with other libraries like NumPy, Matplotlib, and Scikit-Learn.

## 3 Understanding Series & DataFrame
- **Series**: One-dimensional labeled array, like a column in Excel.
- **DataFrame**: Two-dimensional table, similar to an Excel sheet or SQL table.

### Creating a Series from Python Lists or Dictionaries

````python
marks = [30,40,50,60,70]
marks = pd.Series([30,40,50,60,70],index = ['Ronaldo','Messi','Virat','Dhoni','Mbappe'])
marks
````

### Creating a DataFrame

````python
info = pd.DataFrame(
 {'name':['Yash','Rahul','Ravi','Shyam'],
 'age':[19,21,22,23],
 'marks':[100,80,70,85]})
print(info)
print(info['age'])
#Selecting a column
print(info['name'])
#To get specific row we use.loc
print([info.loc[1]])
#Select a specific value
print(info.loc[1,'name'])
````

### Creating DataFrame from NumPy Arrays

````python
info = pd.DataFrame([
    ['Rahul',52,80],
    ['Devraj',22,85],
    ['Abhishek',60,55]
],columns = ['Name','Age','Marks'])
print(info)
print(info.loc[2,'Marks'])
print(info.loc[2])
````

````python
#Generate a 3X4 matrix with random numbers(1-13) using numpy
#Name the column as [A,B,C,D]
import numpy as np
arr = np.random.randint(1,13,(3,4))
arr
arr1 = pd.DataFrame(arr,columns = ['A','B','C','D'])
print(arr1)
````
