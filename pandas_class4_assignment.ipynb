# Q1. Select Single Column

## Problem Statement
Create a DataFrame and display only the Name column.

## Sample Input
```python
Name, Age
```

## Sample Output
```python
Only Name column
```


````python
import pandas as pd
a = {
    'Name':['Virat','Rahul','Shreyas','Yashashvi'],
    'Age':[31,30,29,24],
}
b = pd.DataFrame(a)
print(b)
print(b['Name'])
````


# Q2. Select Multiple Columns

```python
Name, Marks
```

```python
Selected columns
```


# Q7. Sort Data

## Problem Statement
Sort students by Marks.

## Sample Input
```python
Marks column
```

## Sample Output
```python
Sorted DataFrame
```


````python
a = pd.DataFrame({
    'Name':['Aakash','Aashu','Shasank','Shahid'],
    'Marks':[98,56,43,67],
    'City': ['Delhi','Agra','Varanasi','Chandigarh'],
})
print(a)
print(a.sort_values(by = 'Marks'))
````


# Q9. Group By City

## Problem Statement
Find average marks city-wise.

## Sample Input
```python
City and Marks
```

## Sample Output
```python
Grouped result
```


````python
a = pd.DataFrame({
    'Name':['Aakash','Aashu','Shasank','Shahid'],
    'Marks':[98,56,43,67],
    'City': ['Delhi','Agra','Varanasi','Chandigarh'],
    'Age':[19,20,21,22],
})
print(a.groupby('City')['Marks'].mean())
````


# Q10. Multiple Conditions

## Problem Statement
Display students with Marks > 70 and Age > 20.

## Sample Input
```python
Marks and Age
```

## Sample Output
```python
Filtered rows
```


````python
filtered_rows = a[(a['Marks'] > 70) & (a['Age'] > 20)]

print(filtered_rows)
````
